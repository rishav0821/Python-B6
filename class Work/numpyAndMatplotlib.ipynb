```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Create a simple line plot
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, color='red', marker='o', linestyle='--', linewidth=2)
plt.title("Customized Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)  # Adds a grid
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [30, 25, 20, 15]

# Create a 1x2 grid of subplots
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(x, y1, color='blue')
plt.title("Plot 1")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(x, y2, color='green')
plt.title("Plot 2")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x and y values for the linear graph
x = np.linspace(-10, 10, 100)  # 100 evenly spaced points from -10 to 10
y = 2 * x + 3  # Linear equation y = 2x + 3

# Plot the graph
plt.plot(x, y, label='y = 2x + 3')

# Add labels, title, and legend
plt.title("Linear Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np  # Import numpy

pi = 22 / 7  # You can also use np.pi for more precision
omega = 10
X = np.linspace(0, 1, 100)  # Define X as an array (e.g., 100 points between 0 and 1)
Y = np.sin(2 * pi * omega * X)  # Calculate Y
print("Y.ndim:", Y.ndim)  # Check the number of dimensions of Y
print(Y[0:10])
```

```python
plt.plot(X,Y)
plt.title("sine ")
plt.show()
```
